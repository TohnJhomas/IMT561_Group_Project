In this document we're attempting to display the calls on a map colored by type. 

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import folium
import contextily as ctx

# Load the dataset
data = pd.read_csv('./data/cleaned_real_time_fire_2025.csv')
data.head()
```

```python
# convert to geoadataframe
gdf = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data['Longitude'], data['Latitude']))
gdf.crs = "EPSG:4326" # set the coordinate reference system to WGS84
gdf.head()
```

```python
gdf.plot(markersize=1, figsize=(10, 10), column="Type", legend=True)
ctx.add_basemap(plt.gca(), crs=gdf.crs.to_string())
plt.show()
```
